## DICTIONARY
- Python dictionary is a data structure that stores information in key-value pairs.
- While keys must be unique and immutable (like strings or numbers)
- Values can be of any data type, whether mutable or immutable.

```python
data = { "name": "sourav", "age": 22 }
print(data)
print(type(data))
```

```python
#creating dictionary
a = {1: 'Geeks', 2: 'For', 3: 'Geeks'}
print(a)

b = dict(a = "Geeks", b = "for", c = "Geeks")
print(b)
```

```python
#Accessing items
d = { "name": "SOURAV", 1: "Python", (1, 2): [1,2,4] }

# Access using key
print(d["name"])
print(d[1])

# # Access using get()
# print(d.get("name"))
d.get('fruits', 'not_found')
print(d.get("fruits"))
```

```python
#Iterating Through a Dictionary
#you can use loops to iterate over dictionaries ,keys values or items
dict = {1: 'Geeks', 2: 'For', 'age':22}

# Iterate over keys
for key in dict.keys():
    print(key)

# Iterate over values
for value in dict.values():
    print(value)

# # Iterate over key-value pairs
for key, value in dict.items():
    print(f"{key}: {value}")
```

```python
students={
    "student1":{"name":"sourav","age":22},
    "student2":{"name":"kanan","age":19}
}
```

```python
print(students["student1"]["name"])
print(students["student1"]["age"])
```

```python
#iterating over nested dictionaries
for student_id,student_info in students.items():
    print(student_id,student_info)
    for key,value in student_info.items():
        print(key,value)
```

```python
#creating dictionary with dict comprehension
squares = {x: x*2 for x in range(6)}
print(squares)


#x takes values from 1 to 5 and x becomes the key
#x**2 becomes the value
#Each key and value pair is added to the dictionary in one line


even_sq={x:x**2 for x in range(10) if x%2==0}
print(even_sq)
```

```python
#using zip() to create a dictionary from two lists
keys = ['a','b','c','d','e']
values = [1, 2, 3, 4, 5]     

d = {k:v for (k,v) in zip(keys, values)}  
print (d)
```

```python
# using fromkeys() to create a dictionary from a list of keys
d = dict.fromkeys(range(5), True)
print(d)
```

```python
#merge dictionaries

d1={"a":1,"b":2}
d2={"c":3,"d":4}
# d3=d1+d2
# print(d3)
d3={**d1,**d2}
print(d3)
```

```python
# setdefault method
student={"name":"Sourav","age":22}
# student.setdefault("city","Delhi")  # adds only if key doesn't exist
# print(student)
student.setdefault("name","Rahul")  # won't change because key exists
print(student)
```

```python
# make dictionary with user input
d = {}

n = int(input("How many entries: "))

for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    d[key] = value

print(d)
```
