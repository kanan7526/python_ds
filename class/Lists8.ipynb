# List Comprehension

```python
## syntax

## basic syntax [expression for item in iterable]
#with conditional logic [expression for item in iterable if condition]
#nested list comprehension [expression for item1 in iterable for item2 in iterable]
```

```python
#this is long task to do so we use list comprehension
lst=[]
for i in range(2,10):
    if i%2==0:
            lst.append(i)

print(lst)



```

```python
lst= [i**2 for i in range(0,10)]
lst
```

```python
n=[i for i in range(0,20) if i%2==0]
print(n)


# if we want to use if in list comprehension we use it at last
```

```python
eo=[i if i%2==0 else i**2 for i in range(0,20)]
# print(eo)



#if we want to use if else we write it at starting
```

```python
#nested list comprehension


l1=[2,3,4,5,6]
l2=[4,5,8,9,1]

#output - [(4,4),(5,5)]  or [4,5] 
# 1st method 
ans=[(i,j) for i in l1 for j in l2]
# common_elements=[(i,i) for i in l1 if i in l2]
# ans= [(i,j) for i in l1 for j in l2 if i==j]
print(ans)
```

```python
# list comprehension with  built-in function
words = ["hello", "world", "python", "programming"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
#converting string to list
s = "Geeks for Geeks"

a = s.split()
print(a)


#to convert list to string we use join method
str1 = " ".join(a)
print(str1) 



#to string into list of individual characters then we can use list()
str2 = "Hello"
char_list = list(str2)
print(char_list)





#using list comprehension to convert string to list of characters
s = "Geeks for Geeks"
a = [ch for ch in s]
print(a)
```

```python
# Remove duplicates from a list
nums=[1,2,3,2,4,3,5,1]
unique=[]
for num in nums:
    if num not in unique:
        unique.append(num)
print("Without duplicates:",unique)
```
