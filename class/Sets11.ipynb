# Methods in sets

```python

setA={1,2,3,4,5}
setB={4,5,6,7,8}
```

```python
# 1. Union

setC= setA.union(setB)
print("Union:",setC)
#all unique (different elements print)
```

```python

# 2. Intersection

setD= setA.intersection(setB)
print("Intersection:",setD) 
#common elements
```

```python
# 3. Difference

# setE= setA.difference(setB)
set3=setB.difference(setA)
print(set3)
# print("Difference:",setE)

# opposite of intersection
```

```python
# 4. Symmetric Difference

setF= setA.symmetric_difference(setB)
print("Symmetric Difference:",setF)
#common elements are ignored
```

```python
# 5. Subset

setG={1,2}
is_subset= setG.issubset(setA)
print("Is Subset:",is_subset)
#checks set g is present in set a or not
```

```python
# 6. Superset
setG={4,5}
is_superset= setA.issuperset(setG)
print("Is Superset:",is_superset)
#all set g elemets are present in set a
```

```python
# 7. Disjoint
setH={4,10} 
setA={1,2,3}
is_disjoint= setA.isdisjoint(setH)
print("Is Disjoint:",is_disjoint)
#checks that setH elemts are not present in setA
```
