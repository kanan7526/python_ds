## Tuples

- They are same as lists
- but immutable nature makes them different
- Can hold elements of different data types.
- These are ordered, heterogeneous and immutable


```python
# Creating a Tuple

tup = ()
print(tup)

# Using String
tup = ('Geeks', 'For')
print(tup)

# without comma it becomes a string
not_tuple=("apple")
print(type(not_tuple))  # <class 'str'>

# Using List
li = [1, 2, 4, 5, 6]
print(tuple(li))
li = tuple([1, 2, 4, 5, 6])
print(type(li))

# Using Built-in Function
tup = tuple('Geeks')
print(tup)
```

```python
tup = (5, 'Welcome', 7.5, True, [1, 2, 3], {'key': 'value'})
print(tup)
```

```python
#packing and unpacking of tuples

# #packing
a = 1, 2, 3
print(a)

# # Tuple unpacking
tup = ("Geeks", "For", "Geeks")

# This line unpack values of Tuple1
a, b, c = tup
print(a)
print(b)
print(c)

#Tuple Unpacking with Asterisk (*)
# tup = (1, 2, 3, 4, 5)
# a, *b, c ,d= tup
# print(a) 
# print(b) 
# print(c)
# print(d)
```

```python
#accesing tuples

tup = tuple("Sourav")
print(tup[0])
print(tup[1:4])  
print(tup[:3])
```

```python
tup1 = (1, 2, 3,1,1)
```

```python
tup1 * 3
```

```python
# conacatenation of tuples
tup1 = (1, 2, 3)
tup2 = ('Geeks', 'For', 'Geeks')
tup3 = tup1 + tup2
print(tup3)
```

```python
#to modify strings we have to convert it into list and then modify it and then convert it back to string
t1=(10,20,30,40,50)
t1= list(t1)
t1[3]=100
t1.insert(2,350)
t1=tuple(t1)
print(t1)
```

```python
#tuple methods

print(tup1.count(1))
print(tup1.index(2))
```

```python
#nested tuples
#nested list
nested_list = [[1, 2,3],[4,5,6],[7,8,9]]
nested_list[0][2]

# nested_tuple=[[1, 2,3],[4,5,6],(7,8,9)]
# nested_tuple[2][2]
```

```python
nested_tuple2=((1,2,3),("hello", "how","are"),(True,False))
nested_tuple2[0][1]
```

```python
#iterating in tuples
for sub_tup in nested_tuple2:
    for item in sub_tup:
        print(item)
```

```python
# tuple concatenation
t1=(1,2,3)
t2=(4,5,6)
t3=t1+t2
print(t3)   # (1, 2, 3, 4, 5, 6)
```

```python
# when to use tuples over lists?
# 1. when you want data to be constant (not changeable)
# 2. tuples are faster than lists
# 3. tuples can be used as dictionary keys, lists cannot
```
