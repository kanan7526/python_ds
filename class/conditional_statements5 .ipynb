```python
#if the statement is true ,execute the block of code inside the if statement
```

```python
age = 1
if age>=18: print("Eligible to vote.") 
```

```python
#else
#else statement is executed when the condition in the if statement is false
```

```python
age = 16
if age >= 18:
    print("Eligible to vote.")
else:
    print("Not eligible to vote.")
```

```python
#elif
#elif statement is used to check mutiple conditions. 
# It is used when there are more than two possible outcomes.
```

```python
age = 2100

if age <= 12:
    print("Child.")
    
elif age <= 19:
    print("Teenager.")
elif age <= 35:
    print("Young adult.")
else:
    print("Adult.")
```

```python
#nested conditional statements
#Nested conditional statements are if statements that are contained within another if ,elif or else statement.
```

```python
num=int(input("Enter a number: "))
if num >0:
    print("Positive number.")
    if num % 2 == 0:
        print("Even number.")
    else:
        print("Odd number.")
else:
    print("Negative number or zero.")
```
