```python
#basic data types
# The standard built-in data types are categorized as follows:
# 1. Numeric Types
# Integer (int): Represents positive or negative whole numbers without decimals (e.g., 10, -5). In Python 3, integers have unlimited precision and are only limited by available memory.

# Floating-Point (float): Represents real numbers with decimal points or scientific notation (e.g., 3.14, 1.5e2).

# Complex (complex): Represents numbers with a real and an imaginary part, written in the form a + bj (e.g., 2 + 3j). 

# 2. Sequence Types

# String (str): An immutable sequence of Unicode characters enclosed in single, double, or triple quotes. Python does not have a separate "character" type; a character is simply a string of length one.

# List (list): A mutable, ordered collection of items that can be of different data types (e.g., [1, "apple", 3.5]).

# Tuple (tuple): Similar to a list but immutable, meaning it cannot be changed after creation (e.g., (10, 20)).

# Range (range): Represents an immutable sequence of numbers, commonly used for looping a specific number of times. 

# 3. Mapping and Set Types

# Dictionary (dict): A mutable collection of key-value pairs, where each unique key maps to a specific value (e.g., {"name": "Alice", "age": 25}).

# Set (set): An unordered collection of unique, immutable items that does not allow duplicates.

# Frozen Set (frozenset): An immutable version of a set. 

# 4. Other Core Types
# # Boolean (bool): Represents one of two logical values: True or False.
```
