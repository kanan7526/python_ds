# PATTERN PRINTING

```python
# Imagine a grid like this:
# (row, column)

# (1,1) (1,2) (1,3)
# (2,1) (2,2) (2,3)
# (3,1) (3,2) (3,3)
# 👉 Your loops are just moving in this grid:
# Outer loop → moves DOWN (row change)
# Inner loop → moves RIGHT (column change)
# 💡 MEMORY:
# 👉 Outer = vertical (down)
# 👉 Inner = horizontal (across)
```

```python
# Row 1 → *
# Row 2 → **
# Row 3 → ***

# 👉 stars = row number
```

```python
# For n = 3
# Row 1 (i = 1)
# Inner loop runs 1 time → *
# Row 2 (i = 2)
# Inner loop runs 2 times → **
# Row 3 (i = 3)
# Inner loop runs 3 times →
```

```python
n = int(input("n"))
for i in range(1, n+1):   # rows
    for j in range(i):    # columns = row number
        print("@", end="")
    print()
```

```python
#INVERTED TRIANGLE
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
# PYRAMID
# 👉 Two things exist:
# Spaces
# Stars

# PART 1: Spaces
# Row 1 → 4 spaces
# Row 2 → 3 spaces
# 👉 spaces = n - row
# PART 2: Stars
# Row 1 → 1
# Row 2 → 3
# Row 3 → 5
# pattern = odd numbers
# 👉 stars = 2×row - 1



n = 5
for i in range(1, n+1):
    print(" " * (n - i), end="")      # spaces
    print("*" * (2*i - 1))            # stars
```

```python
#inverted pyramid
n = 5
for i in range(n, 0, -1):
    print(" " * (n - i), end="")
    print("*" * (2*i - 1))
```

```python
#diamond (combine top pyramids and inverted pyramid)
n = 5

# top
for i in range(1, n+1):
    print(" "*(n-i) + "*"*(2*i-1))

# bottom
for i in range(n-1, 0, -1):
    print(" "*(n-i) + "*"*(2*i-1))
```
