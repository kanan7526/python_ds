```python
n=int(input("enter a number"))
if n<=1:
    print("the number is not a prime")
else:
    prime=True
    for i in range(2,n):
        if n%i==0:
            prime=False
            break
    if prime:
            print(n, "is prime")
    else:
            print(n, "is not prime")
  
```
