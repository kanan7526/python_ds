```python
# Variables
a = 15
b = 4

# Addition
print("Addition:", a + b)  

# Subtraction
print("Subtraction:", a - b) 

# Multiplication
print("Multiplication:", a * b)  

# Division
print("Division:", a / b) 

# Floor Division
print("Floor Division:", a // b)  

# Modulus
print("Modulus:", a % b) 

# Exponentiation
print("Exponentiation:", a ** 2)
```

```python
a = 13
b = 33

print(a > b) #greater than
print(a < b) #less than
print(a == b)  #equals to
print(a != b)   #not equals to
print(a >= b)   #greater than equal to
print(a <= b)   #less than equal to


# output in boolean i.e True or False
```
