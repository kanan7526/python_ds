## String Comparison

```python
# python compares strings character by character using ASCII values
print("apple"=="apple")   # True
print("apple"=="Apple")   # False
print("apple">"banana")   # False -- 'a' comes before 'b'
```

## Practical Examples

```python
# Count vowels

text="Sourav Kanan"
vowels="aeiouAEIOU"
count=0
for char in text:
    if char in vowels:
        count+=1
print("Number of vowels in",text,"is",count)
```

```python
# Reverse words in a sentence

sentence="I love Python"
words=sentence.split(" ")
print(words)
reversed_sentence=" ".join(words[::-1])
print(reversed_sentence)
```
